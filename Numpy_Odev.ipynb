##  PROJE ÖDEVİ  
### **NumPy ile Üniversite Not Analizi Sistemi**

###  Amaç:
Rastgele üretilmiş bir **ders notları veri kümesi** üzerinde NumPy araçlarıyla keşifsel veri analizi (EDA) yapmak.

###  Veri Kümesi Özellikleri:
- 100 öğrenci (`student_id`: 1–100)
- 5 ders: `Matematik`, `Fizik`, `Kimya`, `Bilgisayar`, `İstatistik`
- Her ders için 0–100 arası notlar (tamsayı)
- Notlar normal dağılıma yakın, ortalama=70, std=15 (sınırlar 0–100)

###  Yapılacaklar:
1. **Veri oluşturma**: `np.random.normal()` + uyarlama (clip & round)
2. **Temel istatistikler**: ortalama, medyan, standart sapma, min/max (her ders için)
3. **Başarılı/Başarısız analizi**: Geçme notu 50 → kaç öğrenci kaldı?
4. **En yüksek notu alan öğrenci**: Hangi öğrenci, hangi dersten?
5. **Korelasyon matrisi**: Dersler arası ilişki (isteğe bağlı: `np.cov` ile)
6. **Özel filtreleme**:
   - Tüm derslerden 85+ alan öğrenciler
   - En az bir dersten 30’dan düşük alanlar
7. **Görselleştirme (ekstra)**: `matplotlib` ile histogram (isteğe bağlı)

###  Teslimat:
- Tek bir `.ipynb` dosyası
- Her adım açıkça yorum satırlarıyla açıklansın
- Kod temiz, okunabilir ve verimli olsun (`for` döngüsü yerine vektörel işlemler tercih edilsin!)

###  Bonus:
- Not dağılımını normalize edip z-skorlarını hesaplayarak "standart sapma üstü/altında" analizi yapın.
- Her öğrencinin genel ortalamasını hesaplayıp, sınıf sıralamasını çıkarın (`argsort` kullanarak).
